# Random Zoom

```python
import numpy as np
import tensorflow as tf

from matplotlib import pyplot as plt

input_data = np.random.random((32,32))

input_data = np.expand_dims(input_data,axis=2)

print('Input:')
plt.imshow(input_data, interpolation='nearest')
plt.show()

layer = tf.keras.layers.RandomZoom(height_factor=0, width_factor=(-0.5, 0.5),fill_mode='constant')

print('Output:')
fig, axs = plt.subplots(1,4, figsize=(15,15),subplot_kw={'xticks': [], 'yticks': []})
for ax in axs.flat:
    ax.imshow(layer(input_data), interpolation='nearest')
    
plt.show()
```

# Random Contrast

```python
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
import matplotlib as mp

input_data = np.random.random((32,32))

input_data = np.expand_dims(input_data,axis=2)

print('Input:')
plt.imshow(input_data, interpolation='nearest',norm=mp.colors.Normalize(vmin=-1, vmax=1.5))
plt.show()

layer = tf.keras.layers.RandomContrast(factor=(0.2,1.8))

print('Output:')
fig, axs = plt.subplots(1,4, figsize=(15,15),subplot_kw={'xticks': [], 'yticks': []})
for ax in axs.flat:
    ax.imshow(layer(input_data), interpolation='nearest',norm=mp.colors.Normalize(vmin=-1, vmax=1.5))
    
plt.show()
```
