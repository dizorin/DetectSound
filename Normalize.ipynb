```python
import numpy as np

#data = np.random.random((128,128))
data1 = np.random.standard_normal((128,128)) * 1 + 0 # (-1 .. 1)
data2 = np.random.standard_normal((128,128)) / 2 + 0.5 # (0 .. 1)
data3 = np.random.standard_normal((128,128)) * 2 + 0 # (-2 .. 2)

data1 = np.expand_dims(data1,axis=2)
data2 = np.expand_dims(data2,axis=2)
data3 = np.expand_dims(data3,axis=2)

data = np.concatenate((data1,data2,data3),axis=2)
print(data.shape)

for i in range(3):
    print("Slice ", i)
    print("In:", data[..., i].mean(), data[..., i].std())
    
    data[..., i] = data[..., i] / np.std(data[..., i])
    data[..., i] = data[..., i] - np.mean(data[..., i])

    print("Out:", round(data[..., i].mean(), 3), round(data[..., i].std(), 3))
    
print("All slice", round(data.mean(),3),round(data.std(),3))
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

data1 = np.random.standard_normal((128,128)) * 1 + 0 # (-1 .. 1)
data2 = np.random.standard_normal((128,128)) / 2 + 0.5 # (0 .. 1)
data3 = np.random.standard_normal((128,128)) * 2 + 0 # (-2 .. 2)

data1 = np.expand_dims(data1,axis=2)
data2 = np.expand_dims(data2,axis=2)
data3 = np.expand_dims(data3,axis=2)

data = np.concatenate((data1,data2,data3),axis=2)
print(data.shape)

scaler = MinMaxScaler()

for i in range(3):
    print("Slice ", i)
    print("In:", data[..., i].min(), data[..., i].max())
    
    data[..., i] = scaler.fit_transform(data[..., i])

    print("Out:", round(data[..., i].min(), 3), round(data[..., i].max(), 3))
    
print("All slice", round(data.min(),3),round(data.max(),3))
```
